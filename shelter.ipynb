## Using Seaborn's Object Model in combination with Ax model 


```python
import matplotlib.pyplot as plt
import seaborn.objects as so 
counts = df['Animal Type'].value_counts()

#Create the Seaborn plot object 
p = so.Plot(df, x = 'Animal Type').add(so.Bar(), so.Hist())
p.label(title = 'Total Animals by Type')
#Create the standard ax plot
ax = sns.countplot(df, x = 'Animal Type')
ax.bar_label(ax.containers[0])
ax.set_title('Total animals by Type')

#Use Seaborn's .on method to combine the two! 
p.on(ax).show()
```

```python
import re 
breeds = dogs['Breed'].to_list()
new_breeds = []
for breed in breeds:
    match = re.findall(r'^Pit', breed)
    if len(match) > 0:
        new_breeds.append('Bully')
    else:
        new_breeds.append(breed)
```

```python
f, ax = plt.subplots()

ax = sns.countplot(bullies, x = 'Intake Condition')
ax.bar_label(ax.containers[0])
ax.tick_params(axis = 'x', labelrotation = 35)
```

```python
nonbully = dogs[~dogs['Breed'].isin(['Bully'])]
nonbully.head()
```

```python
bully_percentages = []
for val in bullies['Intake Condition'].value_counts():
    percentage = (val/len(bullies)) * 100
    bully_percentages.append(percentage)
    
other_percentages = []
for val in nonbully['Intake Condition'].value_counts():
    percentage = (val/len(nonbully)) * 100
    other_percentages.append(percentage)
```

```python
bully_intake_labels = ['Normal', 'Injured', 'Sick','Nursing','Other','Medical','Neonatal','Pregnant','Aged','Behavior','Med Attn','Unknown','Parvo']
```

```python
nonbully['Intake Condition'].value_counts()
```

```python
non_bully_labels = ['Normal','Injured','Sick','Nursing','Aged','Neonatal','Other','Medical','Pregnant','Behavior','Feral','Med Attn','Unknown','Med Urgent','Neurologic','Agonal']
```

```python
f, ax = plt.subplots(figsize = (15,15))
explode = [0.3] * 13
ax = plt.pie(bully_percentages, 
             labels = bully_intake_labels,
             autopct= '%1.3f%%',
             explode= explode
             )
plt.title('Percentages of Bullies by Intake Condition')
f.show()
```

```python
#How does this compare to nonbully breeds 
#Make a donut pie chart with both??


fig, ax = plt.subplots(figsize = (20,20))
ax.axis('equal')
width = 0.3


cm = plt.get_cmap("tab20c")
cout = cm(np.arange(3)*4)
pie, _ = ax.pie(bully_percentages, radius=1, labels= bully_intake_labels, colors=cout)
plt.setp( pie, width=width, edgecolor='white')

cin = cm(np.array([1,2,5,6,9,10]))
pie2, _ = ax.pie(other_percentages, radius=1-width, labels=non_bully_labels,
                                      labeldistance=0.7, colors=cin)
plt.setp( pie2, width=width, edgecolor='white')
plt.show()
```

```python
#How does this compare to nonbully breeds 
#Make a donut pie chart with both??


fig, ax = plt.subplots(figsize = (20,20))
ax.axis('equal')
width = 0.3

wedges, texts = ax.pie(bully_percentages, wedgeprops=dict(width=0.5), startangle=-40)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = f"angle,angleA=0,angleB={ang}"
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(bully_intake_labels[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)


plt.show()

#pie, _ = ax.pie(bully_percentages, radius=1, labels= bully_intake_labels)
#plt.setp( pie, width=width, edgecolor='white')

#pie2, _ = ax.pie(other_percentages, radius=1-width, labels=non_bully_labels,labeldistance=0.7)
#plt.setp( pie2, width=width, edgecolor='white')
#plt.show()
```
